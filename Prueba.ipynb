```python
import pandas as pd
```

```python
data1 = pd.read_excel("datos.xlsx", sheet_name="Captura de imagnes")
data2 = pd.read_excel("datos.xlsx", sheet_name="Tilling")
```

```python
data1["Image_Name"] = data1["Image_Name"].astype(str).str.replace(r'(\Wjpg)', r'', regex=True)
data2["Image_Name"] = data2["Image_Name"].astype(str).str.replace(r'(\W\d[\d]?[\d]?\W\Wjpg)', r'', regex=True)
```
